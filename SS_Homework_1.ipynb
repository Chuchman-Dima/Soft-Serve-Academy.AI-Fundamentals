 Rock-Paper-Scissors game

```python
import numpy as np
from numpy import random

def play_rock_paper_scissors(player_choice, comp_choice):
    if player_choice == comp_choice:
        return 'Tie'

    win = [
        player_choice == 'rock' and comp_choice == 'scissors',
        player_choice == 'scissors' and comp_choice == 'paper',
        player_choice == 'paper' and comp_choice == 'rock',
    ]

    if any(win):
        return 'PLayer 1 wins'
    else:
        return "Comp wins"

def computer_makes_choice():
    valid_choices = ['rock', 'paper', 'scissors']
    return random.choice(valid_choices)


def play_multiple_rounds(num_rounds):
    valid_choices = ['rock', 'paper', 'scissors']

    for i in range(1, num_rounds + 1):
        player_choice = input("Enter your choice: ").lower()

        if player_choice not in valid_choices:
            print("Invalid choice")
            continue

        comp_choice = computer_makes_choice()
        result = play_rock_paper_scissors(player_choice, comp_choice)
        print(f"Round {i}: You chose {player_choice}. Computer chose {comp_choice}. {result}\n")

if __name__ == '__main__':

    num_rounds = int(input("Enter number of rounds: "))
    play_multiple_rounds(num_rounds)
```

Hangman game

```python
import random

def select_random_word():
    word_list = ["apple", "banana", "cherry", "dog", "elephant", "flower", "giraffe", "hamburger", "icecream", "jacket"]
    return random.choice(word_list)

def display_word(word, guessed_letters):
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter
        else:
            display += "_"
    return display

def hangman_game():
    word_to_guess = select_random_word()
    guessed_letters = []
    max_attempts = 6
    incorrect_guesses = 0

    print("Welcome to Hangman!")
    print("Try to guess the word. You can make up to 7 wrong guesses.")

    while incorrect_guesses < max_attempts:
        current_guess = display_word(word_to_guess, guessed_letters)
        print(f"\nWord:{current_guess}")

        if "_" not in current_guess:
            print("You win!")
            break

        guess = input("Enter your guess: ").lower()

        if guess in guessed_letters:
            print(f"You already tried '{guess}'!")
            continue

        guessed_letters.append(guess)

        if guess in word_to_guess:
            print("Good job! You found a letter!")
        else:
            print("You missed!")
            incorrect_guesses += 1
            print(f"Attempts left: {max_attempts - incorrect_guesses}")
    else:
        print("Game over! The word was " + word_to_guess)


if __name__ == "__main__":
    hangman_game()
```
